```python
# Import necessary libraries
import matplotlib.pyplot as plt

# 1. Creating data and storing them in lists
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
z = [1, 4, 9, 16, 25, 36]

# 2. Plotting data in lists
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = 2x', marker='o')
plt.plot(x, z, label='z = x^2', marker='s')
plt.title("Plotting Lists Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# 3. Initializing and accessing list elements with indexing
print("Accessing Elements with Indexing:")
print(f"First element of x: {x[0]}")
print(f"Last element of y: {y[-1]}")
print(f"Third element of z: {z[2]}")

# 4. List slicing, striding, and list operations
print("\nList Slicing and Striding:")
print(f"First three elements of x: {x[:3]}")
print(f"Every second element of y: {y[::2]}")
print(f"Reversed z: {z[::-1]}")

# List operations
list_sum = [x[i] + y[i] for i in range(len(x))]
print("\nList Operations:")
print(f"Sum of x and y element-wise: {list_sum}")

# 5. Looping over a list with for
print("\nLooping Over a List:")
for i, value in enumerate(x):
    print(f"Element at index {i} in x: {value}")
```
